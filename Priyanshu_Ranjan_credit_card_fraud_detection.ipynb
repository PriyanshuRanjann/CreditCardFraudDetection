```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt  
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
```

```python
data=pd.read_csv("creditcard.csv")
```

```python
#legit and fraud transaction
data['Class'].value_counts()
```

```python
#0 represents normal transaction and 1 represents fraudulent transaction.
#The distribution of both data point is unbalance as more than 99% of data belongs to one type .Hence processing is done
```

```python
#seperating data for analysis
normal= data[data.Class==0]
fraud= data[data.Class==1]
```

```python
print("Shape of real transaction:",normal.shape)
print("Shape of Fraud Transaction:",fraud.shape)
```

```python
#Statistical analysis
normal.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
data.groupby('Class').mean()
```

```python
#Under-sampling-Build a sample dataset containing similar distribution 
#Number of fraud transaction--> 492
```

```python
normal_sample=normal.sample(n=492)
```

```python
#concatenate the datas
```

```python
newdata=pd.concat([normal_sample,fraud],axis=0)
```

```python
newdata.head()
```

```python
newdata['Class'].value_counts()
```

```python
newdata.groupby('Class').mean()
```

```python
x=newdata.drop(columns='Class',axis=1)
y=newdata['Class']
```

```python
# Splitting data into train and test dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#Model training
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(x_train,y_train)
```

```python
#Evaluation
#Acuracy score
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
print('Accuracy on training data:',training_data_accuracy)
```

```python
#Accuracy on test data
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
print('Accuracy on test data:',test_data_accuracy)
```

```python
# Accuracy on both training and testing data are similar there is no case of underfitting and overfitting.
```

```python
x_train_prediction=model.predict(x_train)
training_data_f1=f1_score(x_train_prediction,y_train)
print('F1 score on training data:',training_data_accuracy)
#F1 score on test data
x_test_prediction=model.predict(x_test)
test_data_F1=f1_score(x_test_prediction,y_test)
print('F1 Score on test data:',test_data_accuracy)
```

```python
precision_train=precision_score(x_train_prediction,y_train)
print('Precision Score on train data:',precision_train)

precision_test=precision_score(x_test_prediction,y_test)
print('Precision Score on test data:',precision_test)
```

```python
recall_train=recall_score(x_train_prediction,y_train)
recall_test=recall_score(x_test_prediction,y_test)
print('Recall Score on train data:',recall_train)
print('Recall Score on test data:',recall_test)
```
